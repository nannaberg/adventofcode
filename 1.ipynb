```python
data_path = "/content/drive/MyDrive/adventofcode/2020/data_1.txt"
with open(data_path) as f:
    report = [int(word) for word in f.read().splitlines()]
```

```python
def get_2_numbers(report, target):
    report.sort()
    t = target
    i = 0
    j = len(report)-1
    while i < j:
        temp_sum = report[i] + report[j]
        if temp_sum == t:
            return (True, report[i], report[j])
        elif temp_sum < t:
            i+=1
        else:
            j-=1
    return (False, report[i], report[j])
            
testing = get_2_numbers(report[:100], 2020)
print(testing)
print(testing[1]*testing[2])
```

```python
def get_3_numbers(report):
    report.sort()
    t = 2020
    i = 0
    j = 1
    k = len(report)-1
    while k > 1:
        bla = 2020-report[k]
        res = get_2_numbers(report[:k],bla)
        if res[0]== False:
            k-=1
        else:
            final_sum = res[1] + res[2] + report[k]
            return final_sum, res[1], res[2], report[k]

foo = get_3_numbers(report)
print(foo)
print(foo[1]*foo[2]*foo[3])
```
