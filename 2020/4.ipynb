```python
import re
from itertools import groupby
```

```python
data_path = "/content/drive/MyDrive/adventofcode/2020/data_4.txt"

def parse_file(data_path):
    with open(data_path) as f:
        lines = f.read().splitlines()
        passports = [" ".join(list(sub)) for ele, sub in groupby(lines, key = bool) if ele]
        return passports

passports = parse_file(data_path)
print(len(passports))
print(passports[:5])

req_fields = ["byr", "iyr" , "eyr", "hgt", "hcl", "ecl", "pid"]
req_field_formats = ["200[0-2]|19[2-9][0-9]$", "^201[0-9]|2020$", "^202[0-9]|2030$", "^1([5-8][0-9]|9[0-3])cm|([5-6][0-9]|7[0-6])in$",
                         "^#([a-f]|\d){6}$", "^(amb|blu|brn|gry|grn|hzl|oth){1}$", "^0*\d{9}$"]
format_dict = dict(zip(req_fields, req_field_formats))
print(format_dict)
```

```python
def part_1(passports):
    valid_counter = 0
    for passport in passports:
        if all(field in passport for field in req_fields):
            valid_counter += 1
    return valid_counter

print(part_1(passports))
```

```python
def part_2(passports):
    valid_counter = 0
    for passport in passports:
        if all(field in passport for field in req_fields):
            field_dict = {item[0]: item[1] for item in [re.split(":", ele) for ele in passport.split()]}
            count = 1
            for field in req_fields:
                if not re.match(format_dict[field], field_dict[field]):
                    count = 0
                    break
            valid_counter += count
    return valid_counter

print(part_2(passports))
```
