```python
import numpy as np
```

```python
data_path = "/content/drive/MyDrive/adventofcode/2020/data_5.txt"

def parse_file(data_path):
    with open(data_path) as f:
        passes = f.read().splitlines()
        #passports = [" ".join(list(sub)) for ele, sub in groupby(lines, key = bool) if ele]
        return passes

passes = parse_file(data_path)
print(passes)
print(len(passes))
```

```python
b_pass = "FBFBBFFRLR"

def get_lower_range(some_range):
    start = some_range[0]
    stop = start + int(len(some_range)/2)
    return np.arange(start, stop)

def get_upper_range(some_range):
    start = some_range[int(len(some_range)/2)]
    stop = some_range[-1]+1
    return np.arange(start, stop)

def get_row(b_pass):
    rows = np.arange(128)
    for char in b_pass[:7]:
        if char=="F":
            rows = get_lower_range(rows)
        else:
            rows = get_upper_range(rows)
    return rows[0]

print(get_row(b_pass))

def get_col(b_pass):
    cols = np.arange(8)
    for char in b_pass[-3:]:
        if char == "L":
            cols = get_lower_range(cols)
        else:
            cols = get_upper_range(cols)
    return cols[0]

def calculate_id(b_pass):
    return get_row(b_pass)*8 + get_col(b_pass)


print(calculate_id(b_pass))
```

```python
ids = [calculate_id(b_pass) for b_pass in passes]
print(ids)
print(len(ids))
print(max(ids))
```

```python
ids.sort()
print(ids)

prev = ids[0]
for id in ids[1:]:
    if id - prev != 1:
        print(prev+1)
    prev = id
```
