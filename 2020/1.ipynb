```python
import numpy as np
```

```python
data_path = "/content/drive/MyDrive/adventofcode/2020/data_1.txt"
with open(data_path) as f:
    report = [int(word) for word in f.read().splitlines()]
```

```python
#part 1
def product_2_entries(report, target):
    report.sort()
    i = 0
    j = len(report)-1
    while i < j:
        temp_sum = report[i] + report[j]
        if temp_sum == target:
            return report[i] * report[j]
        elif temp_sum < target:
            i+=1
        else:
            j-=1
            
print("Answer part 1: ", product_2_entries(report, 2020))
```

```python
#part 2
def product_3_entries(report,target):
    report.sort()
    k = len(report)-1
    while k > 1:
        new_target = target-report[k]
        prod_2 = product_2_entries(report[:k],new_target)
        if prod_2==None:
            k-=1
        else:
            return prod_2 * report[k]

print("Answer part 2: ", product_3_entries(report,2020))
```
