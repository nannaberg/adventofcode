```python
import numpy as np
import math
import time
```

```python
data_path = "/content/drive/MyDrive/adventofcode/2021/data_9.txt"

def parse_file(data_path):
    with open(data_path) as f:
        lines = [[int(number) for number in word] for word in f.read().splitlines()]
    return lines

smoke_map = parse_file(data_path)
print(len(smoke_map))
print(len(smoke_map[0]))
print(smoke_map)
```

```python
def pad_smoke_map(smoke_map):
    smoke_map = np.array(smoke_map)
    def pad_with(vector, pad_width, iaxis, kwargs):
        pad_value = kwargs.get('padder', 10)
        vector[:pad_width[0]] = pad_value
        vector[-pad_width[1]:] = pad_value

    padded_smoke_map = np.pad(smoke_map, 1, pad_with, padder=100)
    return padded_smoke_map

def risk_level_sum(padded_smoke_map):
    min_array = np.zeros_like(padded_smoke_map)

    for i in range(padded_smoke_map.shape[0]):
        min_array[i,1:-1] += np.diff(np.sign(np.diff(padded_smoke_map[i])))

    for i in range(padded_smoke_map.shape[1]):
        min_array[1:-1,i] += np.diff(np.sign(np.diff(padded_smoke_map[:,i])))

    indexes = np.where(min_array == 4)
    minima = padded_smoke_map[indexes]
    risk_level_sum = minima.sum()+len(minima)

    return risk_level_sum

start_time = time.time()
padded_smoke_map = pad_smoke_map(smoke_map)
risk_level_sum = risk_level_sum(padded_smoke_map)
print("Answer part 1 --> risk level sum: ",risk_level_sum)
print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))
```
