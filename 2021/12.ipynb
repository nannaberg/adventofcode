```python
data_path = "/content/drive/MyDrive/adventofcode/2021/data_12.txt"

def parse_file(data_path):
    with open(data_path) as f:
        #lines = [int(number) for line in f.read().splitlines() for number in line]
        lines = f.read().splitlines()
    return lines

data = parse_file(data_path)
print(len(data))
print(data)
```

```python
graph = [tuple(word.split("-")) for word in data] 
print(graph)
```

```python
def get_children(parent, graph):
    children = [word[1] for word in graph if word[0]==parent and word[1] != "start"]
    children.extend([word[0] for word in graph if word[1]==parent])
    return children
print(get_children("start", graph))
print(get_children("dc", graph))

def get_small_caves(graph):
    small_caves = set([char for path in graph for char in path if char not in ["start", "end"] and not char.isupper()])
    return small_caves
print("small_caves:")
print(get_small_caves(graph))
```

```python
path_list = set()

def rec_paths(graph, node, path, visited,small_cave):
    #print(node)
    if node == "end":
        path.append(node)
        if path not in path_list:
            path_list.append(path)
        return None
    elif node in visited:
        return None
            # path.pop()
            # rec_paths(graph, path[-1], path, visited)      
    children = get_children(node, graph)
    #print(children)
    if children:
        if node.islower():
            if node == small_cave:
                small_cave = -1
            else:
                visited.append(node)
        path.append(node)
        for child in children:
            rec_paths(graph, child, path.copy(), visited.copy(),small_cave)
    

small_caves = get_small_caves(graph)
for cave in small_caves:
    print(cave)
    rec_paths(graph,"start",[],[],cave)

print(len(path_list))
# for path in path_list:
#     print(path)
```
