```python
import numpy as np
```

```python
data_path = "/content/drive/MyDrive/adventofcode/2021/data_2.txt"
with open(data_path) as f:
    instructions = [word.split() for word in f.read().splitlines()]
print(instructions[:10])
print(len(instructions))
```

```python
#part 1

pos = 0
depth = 0
for instruction in instructions:
    direction = instruction[0]
    magnitude = int(instruction[1])
    if direction == "forward":
        pos += magnitude
    elif direction == "down":
        depth += magnitude
    else:
        depth -= magnitude
print("position: ",pos)
print("depth: ",depth)
print("Answer part 1 - position * depth: ", pos*depth)
```

```python
#part 2 (and part 1)

pos = 0
depth = 0
aim = 0
for instruction in instructions:
    direction = instruction[0]
    magnitude = int(instruction[1])
    if direction == "forward":
        pos += magnitude
        depth += magnitude * aim
    elif direction == "down":
        aim += magnitude
    else:
        aim -= magnitude
print("position: ", pos)
print("depth: ", depth)
print("aim: ", aim)
print("Answer part 1 - position * aim: ", pos*aim)
print("Answer part 2 - position * depth: ", pos*depth)
```
