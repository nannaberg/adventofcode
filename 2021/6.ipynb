```python
import numpy as np
import time
```

```python
data_path = "/content/drive/MyDrive/adventofcode/2021/data_6.txt"

def parse_file(data_path):
    with open(data_path) as f:
        lines = [int(char) for char in f.read().split(",")]
    return lines
```

```python
def count_fish_aux(day,total_days, memo_table):
    days_left = total_days-day
    if memo_table[days_left] != -1:
        return memo_table[days_left]
    elif days_left < 8:
        return 0
    else:
        days_new_fish_created = np.arange(day+9,total_days,7)
        amount_new_fish_created = len(days_new_fish_created)
        for new_day in days_new_fish_created:
            amount_new_fish_created += count_fish_aux(new_day, total_days, memo_table)
        memo_table[days_left] = amount_new_fish_created
        return memo_table[days_left]

def count_fish(fish_list, total_days):
    counter = len(fish_list)
    memo_table = np.negative(np.ones(total_days,dtype=int))
    
    for fish in fish_list:
        creation_days = np.arange(fish,total_days,7)
        counter += len(creation_days)
        for day in creation_days:
            counter += count_fish_aux(day,total_days, memo_table)
    return counter




days_part_1 = 80
days_part_2 = 256
start_time = time.time()
fish_list = parse_file(data_path)
print("Answer part 1 -- #fish after {} days: ".format(days_part_1), count_fish(fish_list, days_part_1))
print("Answer part 2 -- #fish after {} days: ".format(days_part_2), count_fish(fish_list, days_part_2))
print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))
```
