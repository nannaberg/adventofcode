```python
import numpy as np
import math
import time
```

```python
data_path = "/content/drive/MyDrive/adventofcode/2021/data_10.txt"

def parse_file(data_path):
    with open(data_path) as f:
        lines = f.read().splitlines()
    return lines

navigation = parse_file(data_path)
print(len(navigation))
print(navigation[:3])
```

```python
def find_corrupt_lines(navigation):
    error_list = []
    corrupted_lines = []
    next_closer_list = ["n/a"]

    for line in navigation:
        for char in line:
            if char in ["]", "}", ">", ")"]:
                if char != next_closer_list[-1]:
                    error_list.append(char)
                    corrupted_lines.append(line)
                    break
                else:
                    next_closer_list.pop()
            if char == "[":
                next_closer_list.append("]")
            elif char == "{":
                next_closer_list.append("}")
            elif char == "<":
                next_closer_list.append(">")
            elif char == "(":
                next_closer_list.append(")")
    return error_list, corrupted_lines


start_time = time.time()

error_list, corrupted_lines = find_corrupt_lines(navigation)
score_dict = {")":3, "]":57, "}":1197, ">": 25137}
error_score = sum([score_dict[symbol] for symbol in error_list])
print("Answer part 1 --> error_score: ", error_score)
print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))
```

```python
def get_completion_strings(incomplete_nav):
    completion_string_list = []
    #next_closer_list = ["n/a"]

    for line in incomplete_nav:
        completion_string = []
        #print(line)
        for char in line:
            #print("next closer: ", next_closer_list)
            #print("current char: ", char)
            
            if char in ["]", "}", ">", ")"]:
                completion_string.pop()
            else:
                if char == "[":
                    completion_string.append("]")
                elif char == "{":
                    completion_string.append("}")
                elif char == "<":
                    completion_string.append(">")
                elif char == "(":
                    completion_string.append(")")
        completion_string_list.append(completion_string[::-1])
    return completion_string_list


def string_score(completion_string):
    completion_score_dict = {")":1, "]":2, "}":3, ">": 4}
    total_score = 0
    for char in completion_string:
        total_score *= 5
        total_score += completion_score_dict[char]
    return total_score


start_time = time.time()

incomplete_nav = [line for line in navigation if line not in corrupted_lines]
completion_string_list = get_completion_strings(incomplete_nav)
string_scores = sorted([string_score(completion_string) for completion_string in completion_string_list])
mid_index = int(len(string_scores)/2)
print("Answer part 2 --> middle score: ", string_scores[mid_index])
print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))
```
