```python





def update_rest(seg_posibilities):
    for seg in seg_posibilities[2:]:
        if seg in seg_posibilities[0]:
            seg_posibilities[0] = re.sub(seg, "",seg_posibilities[0])
        elif seg in seg_posibilities[1]:
            seg_posibilities[1] = re.sub(seg, "",seg_posibilities[1])
    return seg_posibilities

def evaluate_output_signals(output_signals, seg_posibilities):
    zero = [0,1,2,3,4,6]
    one = [2,3]
    two = [4,2,5,1,6]
    three = [4,2,5,3,6]
    four = [0,5,2,3]
    five = [4,0,5,3,6]
    six = [4,0,5,1,6,3]
    seven = [4,2,3]
    eight = [0,1,2,3,4,5,6]
    nine = [0,5,2,3,4]
    signal_sum = 0
    for signal in output_signals:
        signal_set = set(signal)
        if signal_set == set(seg_posibilities[(one)]):
            signal_sum += 1
        elif signal_set == set(seg_posibilities[(two)]):
                signal_sum += 2
        elif signal_set == set(seg_posibilities[(three)]):
                signal_sum += 3
        elif signal_set == set(seg_posibilities[(four)]):
                signal_sum += 4
        elif signal_set == set(seg_posibilities[(five)]):
                signal_sum += 5
        elif signal_set == set(seg_posibilities[(6)]):
                signal_sum += 6
        elif signal_set == set(seg_posibilities[(seven)]):
                signal_sum += 7
        elif signal_set == set(seg_posibilities[(eight)]):
                signal_sum += 8
        elif signal_set == set(seg_posibilities[(nine)]):
            signal_sum += 9
    return signal_sum

def update_seg(i,j,reduced_signals,seg_posibilities):
    if reduced_signals[0] in seg_posibilities[i]:
        seg_posibilities[i]=reduced_signals[0]
        seg_posibilities[j]=reduced_signals[1]
    else:
        seg_posibilities[i]=reduced_signals[1]
        seg_posibilities[j]=reduced_signals[0]
    return seg_posibilities

def update_for_2_5_3(seg_posibilities, signals):
    pattern_seg_1 = "["+seg_posibilities[1]+"]"
    pattern_seg_0 = "["+seg_posibilities[0]+"]"
    #print("pattern_seg_1" , pattern_seg_1)
    #print("pattern_seg_1", pattern_seg_0)
    signals_3 = ""
    for signal in signals[3:6]:
        if len(re.findall(pattern_seg_1, signal)) == len(seg_posibilities[1]): #number "2"
            #print("{} is signals for number 2".format(signal))
            pattern = "["+seg_posibilities[4]+seg_posibilities[1]+"]"
            #print(pattern)
            reduced_signals = re.sub(pattern, "", signal)
            #print(reduced_signals)
            #we can now determine two more segment signalss, seg 2 and seg 5
            seg_posibilities = update_seg(2,5,reduced_signals, seg_posibilities)

        elif len(re.findall(pattern_seg_0, signal)) == len(seg_posibilities[0]): #number "5"
            #print("{} is signals for number 5".format(signal))
            pattern = "["+seg_posibilities[4]+seg_posibilities[0]+"]"
            #print(pattern)
            reduced_signals = re.sub(pattern, "", signal)
            #print(reduced_signals)
            seg_posibilities = update_seg(3,6,reduced_signals, seg_posibilities)
        else:
            signals_3 = signals
    return seg_posibilities



def evaluate_output_signals(output_signals, seg_posibilities):
    zero = [0,1,2,3,4,6]
    one = [2,3]
    two = [4,2,5,1,6]
    three = [4,2,5,3,6]
    four = [0,5,2,3]
    five = [4,0,5,3,6]
    six = [4,0,5,1,6,3]
    seven = [4,2,3]
    eight = [0,1,2,3,4,5,6]
    nine = [0,5,2,3,4,6]
    digit_list = []
    for signal in output_signals:
        signal_set = set(signal)
        if signal_set == set(seg_posibilities[(one)]):
            digit_list.append(1)
        elif signal_set == set(seg_posibilities[(two)]):
            digit_list.append(2)
        elif signal_set == set(seg_posibilities[(three)]):
            digit_list.append(3)
        elif signal_set == set(seg_posibilities[(four)]):
            digit_list.append(4)
        elif signal_set == set(seg_posibilities[(five)]):
            digit_list.append(5)
        elif signal_set == set(seg_posibilities[(six)]):
            digit_list.append(6)
        elif signal_set == set(seg_posibilities[(seven)]):
            digit_list.append(7)
        elif signal_set == set(seg_posibilities[(eight)]):
            digit_list.append(8)
        elif signal_set == set(seg_posibilities[(nine)]):
            digit_list.append(9)
        else: 
            digit_list.append(0)
    output_digit = digit_list[0]*1000 + digit_list[1]*100 + digit_list[2]*10 + digit_list[3]
    return output_digit

def reduce_seg_posibilities(signals):
    print("\n in reduce fn")
    seg_posibilities = np.array(["abcdefg","abcdefg","abcdefg","abcdefg","abcdefg","abcdefg","abcdefg"])
    #signals = sorted(signals, key = len)
    #print("signals: ", signals)
    #print(seg_posibilities)
    
    #updating posibilities for segment 2 and 3 which are the segments that create the number "1"
    #print("\nupdating for seg 2 and 3")
    seg_posibilities[[2,3]] = "".join(re.findall("["+signals[0]+"]",seg_posibilities[2]))
    #print(seg_posibilities)


    #print("\nupdating for seg 4")
    #updating posibilities for segment 4 which, along with segments 2 and 3, create the number "7"
    #only one posibility left, namely the char that did not appear in the seg posibilities for segment 2 and 3 (ie number "1")
    seg_posibilities[4] = "".join(re.findall("["+signals[1].strip(seg_posibilities[2])+"]",seg_posibilities[4]))
    #print(seg_posibilities)


    #print("\nupdating for seg 0 and 5")

    #updating posibility for segment 0 and 5, which along with segments 2 and 3, create the number "7"
    seg_posibilities[[0,5]] = "".join(re.findall("["+re.sub("["+seg_posibilities[2]+"]","",signals[2])+"]",seg_posibilities[0]))
    print(seg_posibilities)


    #updating posibility for segment 1 and 6, which along with all the other segments, create the number "8"
    foo = "".join(seg_posibilities[[0,2,4]])
    print(foo)
    print(signals[9])
    bar = "["+re.sub("["+foo+"]","",signals[9])+"]"
    print(bar)
    seg_posibilities[[1,6]] = "".join(re.findall(bar,signals[9]))
    print(seg_posibilities)
    #print(seg_posibilities)
    return seg_posibilities

def do_it_all(lines):
    total_sum = 0
    for line in lines:
        signals = sorted(line[0], key = len)
        print(signals)
        #signals = line[0]
        output_signals = line[1]
        
        seg_posibilities = reduce_seg_posibilities(signals)
        #print(seg_posibilities)
        print(seg_posibilities)
        seg_posibilities = update_for_2_5_3(seg_posibilities, signals)
        print(seg_posibilities)
        seg_posibilities = update_rest(seg_posibilities)
        print(seg_posibilities)
        output_sum = evaluate_output_signals(output_signals, seg_posibilities)
        print("output signals: ", output_signals)
        print("output sum: ", output_sum)
        total_sum += output_sum
    return total_sum

do_it_all(lines)



```
