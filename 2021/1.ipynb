```python
#a more general solution
def depth_increase_counter(report, window_size):
    counter = 0
    prev_sum = sum(report[:window_size])
    for i in range(1,len(report)-(window_size-1)):
        current_sum = sum(input[i:i+window_size])
        if current_sum > prev_sum:
            counter += 1
        prev_sum = current_sum
    return counter

print("Answer part 1: ",depth_increase_counter(report, 1))
print("Answer part 2: ",depth_increase_counter(report, 3))
```
