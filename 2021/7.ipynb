```python
import numpy as np
import time
import math
```

```python
data_path = "/content/drive/MyDrive/adventofcode/2021/data_7.txt"

def parse_file(data_path):
    with open(data_path) as f:
        lines = [int(char) for char in f.read().split(",")]
    return lines

lines = parse_file(data_path)
print(len(lines))
print(lines)
print(max(lines))
print(min(lines))
```

```python
#test_lines = [16,1,2,0,4,2,7,1,2,14]
start_time = time.time()
#pos = 2
fuel_list = []

for pos in range(max(lines)):
    fuel = 0
    for crab in lines:
        fuel += abs(crab-pos) 
    fuel_list.append((fuel,pos))

print(fuel_list)
print(min(fuel_list))
print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))
```

```python
#time using summing over a list

#lines = [16,1,2,0,4,2,7,1,2,14]
start_time = time.time()
#pos = 2
fuel_list = []

for pos in range(max(lines)):
    fuel = 0
    for crab in lines:
        #fuel += np.arange(1,abs(crab-pos)+1).sum()
        fuel += sum(range(1,abs(crab-pos)+1))
    fuel_list.append((fuel,pos))

print(fuel_list)
print(min(fuel_list))
print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))
```

```python
#time using numpy arrays to sum


#lines = [16,1,2,0,4,2,7,1,2,14]
start_time = time.time()
#pos = 2
fuel_list = []

for pos in range(max(lines)):
    fuel = 0
    for crab in lines:
        fuel += np.arange(1,abs(crab-pos)+1).sum()
        #fuel += sum(range(1,abs(crab-pos)+1))
    fuel_list.append((fuel,pos))

print(fuel_list)
print(min(fuel_list))
print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))
```

```python
#using numpy array to sum and discard fuel values the moment they are too big, only saving the current smallest one


#lines = [16,1,2,0,4,2,7,1,2,14]
start_time = time.time()
#pos = 2
#fuel_list = []
min_fuel = (math.inf,math.inf)
for pos in range(max(lines)):
    fuel = 0
    useful = True
    for crab in lines:
        fuel += np.arange(1,abs(crab-pos)+1).sum()
        #fuel += sum(range(1,abs(crab-pos)+1))
        if fuel > min_fuel[0]:
            useful = False
            break
    if useful == True:
        min_fuel = (fuel,pos)
    #fuel_list.append((fuel,pos))

#print(fuel_list)
#print(min(fuel_list))
print(min_fuel)
print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))
```

```python
#using triangle numbers and fuel values as soon as they get too big, only saving the current smallest one


start_time = time.time()
min_fuel = (math.inf,math.inf)
for pos in range(min(lines),max(lines)+1):
    fuel = 0
    for crab in lines:
        n = abs(crab-pos)
        fuel += int(n*(n+1)/2)
        if fuel > min_fuel[0]:
            break
    if min_fuel[0] > fuel:
        min_fuel = (fuel,pos)
print(min_fuel)
print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))
```
